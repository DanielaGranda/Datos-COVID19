```
def TrazaError( x, y ):
  X = sm.add_constant(x)
  model = sm.OLS(y, X).fit()
  m = model.params[1] # pendiente 
  b = model.params[0] # intersecci贸n
  desvStd = np.std(y) # desviaci贸n Estandar
  cCor = np.corrcoef(x,y)[0][1] # Correlaci贸n
  err = desvStd * math.sqrt(1-cCor*cCor) # Error cuadratico
  points = np.linspace(x.min(), x.max())##Limite en el eje X de las rectas de error
  plt.plot(points, m*points + b  )
  plt.plot(points, m*points + ( b + err))
  plt.plot(points, m*points + ( b - err))
  print(err)
```
